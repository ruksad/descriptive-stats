```python
def find_quartile(nums):
    quartiles = [0.25,0.75]
    sorted_nums= np.sort(nums)
    min = np.min(nums)
    max = np.max(nums)
    mean = np.mean(nums)
    total_length = len(nums)

    result = []


    for quartile in quartiles:
        idx = quartile*total_length
        if idx.is_integer():
            idx = int(idx)
            result.append((sorted_nums[idx] + sorted_nums[idx+1])/2)
        else:
            result.append(sorted_nums[np.ceil(idx)])
    return [min, result[0], mean, result[1], max]

quartiles = find_quartile(X)
print(quartiles)
plt.boxplot(quartiles, vert=False)
plt.show()
```

```python
g = np.linspace(1, 10, 200)

```

### Softplus

```python
def softplus(x):
    return np.log(1+ np.exp(x))

fx1 = softplus(g)
```

### Log-Sum-Exp (LSE)

```python
def LSE(nums: np.ndarray):
    return np.log(np.sum(np.exp(nums)))


fx2=LSE(g)
```

### Gaussian Function

```python
def gaussian(x):
    return np.exp((x**2) * (-1/2))

fx3=gaussian(g)
```

### Sigmoid Function

```python
def Sigmoid(x):
    return 1/(1+np.exp(-1*x))

fx4=Sigmoid(g)
```

### Linear-Exponential Function

```python
def LE(x):
    return x*np.exp(-0.1*x)

fx5=LE(g)
```

```python
plt.figure()

fig, axs = plt.subplots(2,2, figsize=(10,10))

axs[0,0].plot(g, fx1, label='Softplus activation')
axs[0,0].set_title('Softplus activation')
axs[0,1].plot(g, fx3, label='Gaussian Function', color = 'red')
axs[0,1].set_title('Gaussian Function')
axs[1,0].plot(g, fx4, label='Sigmoid Function', color='green')
axs[1,0].set_title('Sigmoid Function')
axs[1,1].plot(g, fx5, label='Linear Exponential Function', color='brown')
axs[1,1].set_title('Linear Exponential Function')



plt.show()
```
