```python
# .  Generate a dataset of 100 samples where:
#        • X is a random feature between 1 and 10.
#        • y = 3X + noise (where noise is a small random value).

np.random.seed(42)
X= np.random.uniform(1,10,size=100)

noise= np.random.normal(0,1,100)

Y= 3*X+noise

#print("X:", X)
#print("Y:", Y)
# Convert the dataset into a NumPy array and reshape X into a column vector of shape (100, 1)
reshape_y= Y.reshape(100,1)
print(f"reshape shape :{reshape_y.shape}")

# Normalize X so that it has a mean of 0 and a standard deviation of 1.
xmean= np.mean(X)
std= np.std(X)
standardized_x=  (X-xmean)/ std
print("standardized arr \n ", standardized_x)
print("std mean:", np.mean(standardized_x)) # almost zero
print(" std", np.std(standardized_x)) # 1
standardized_x
```

```python
np.random.seed(42)
x= np.random.uniform(1,10,100)
y= 3*x+np.random.normal(0,1,100)

#standardize x 
mean_x= np.mean(x)
std_x=np.std(x)
stdzd_x= (x-mean_x)/ std_x

print(f"std mean x= {np.mean(stdzd_x)}, mean y={np.mean(y)}")
print(f" std of stdzd_x= {np.std(stdzd_x)}, std y= {np.std(y)}")
# combine x and y
dataset= np.column_stack((stdzd_x,y))

np.random.shuffle(dataset)
# Split the dataset into training (80%) and testing (20%) using NumPy.
split_dataset_index= int(.8* len(dataset))
train_set= dataset[:split_dataset_index]  # starting to .8
test_set= dataset[split_dataset_index:]    #.8 to end

x_train, y_train= train_set[:,0], train_set[:,1]
x_test, y_test= test_set[:,0], test_set[:,1]

print("X-train shape", x_train.shape)
print("Y train-shape", y_train.shape)
print("x_test shape", x_test.shape)
print("y_train shape", y_train.shape)
```

```python
# To compute the correlation coefficient between X and y, you can use the np.corrcoef() function from NumPy:
corrcoef= np.corrcoef(stdzd_x,y) 

print("corrcoef", corrcoef[0,1]) # [[corr(X,X)], corr(x,y)],[cor(y,x), cor(y,y)]] so choose 0,1 index 
# output 0.9935285636853227 strong correlation
```

```python
# Standard Normal Random Variable 
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generate standard normal data
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, 0, 1)  # Standard normal PDF

# Plot the standard normal distribution
plt.plot(x, y, label="Standard Normal PDF")
plt.xlabel("x")
plt.ylabel("Density")
plt.title("Standard Normal Distribution")
plt.legend()
plt.grid()
plt.show()
```
